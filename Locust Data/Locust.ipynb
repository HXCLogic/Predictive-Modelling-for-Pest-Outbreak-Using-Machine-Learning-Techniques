```python
data['Outbreak'] = ((data['HopperBand'] > 0) & (data['Swarms'] > 0)).astype(int)
data.head()
```

```python
count_ones = data['Outbreak'].sum()
total = data['Outbreak'].count()
pct = count_ones / total * 100
print(f"Outbreak == 1: {count_ones}/{total} ({pct:.2f}%)")
```
